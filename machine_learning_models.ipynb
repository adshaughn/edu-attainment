```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import sklearn
```

```python
# Load Data
data = Path('Resources/edu_attain_final.csv')
edu_attain_raw = pd.read_csv(data)
edu_attain_raw
```

```python
# Binary Encooding on asab (gender) column
edu_attain = pd.get_dummies(edu_attain_raw, columns = ["asab"])
edu_attain
```

```python
# Label Encoding on subject column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
edu_attain['subject'] = le.fit_transform(edu_attain['subject'])
edu_attain.head(10)
```

```python
# Get list of all countires
edu_attain['country'].unique().tolist()
```

```python
# Remove OAVG rows
edu_attain = edu_attain[edu_attain['country'] != 'OAVG']
edu_attain
```

```python
# Countries dictionary, excluding OAVG
countries = {
    'AUS': 1,
    'BEL': 2,
    'CAN': 3,
    'CHE': 4,
    'CRI': 5,
    'CZE': 6,
    'DEU': 7,
    'DNK': 8,
    'ESP': 9,
    'EST': 10,
    'FIN': 11,
    'FRA': 12,
    'GBR': 13,
    'GRC': 14,
    'HUN': 15,
    'IRL': 16,
    'ITA': 17,
    'JPN': 18,
    'KOR': 19,
    'LTU': 20,
    'LUX': 21,
    'LVA': 22,
    'MEX': 23,
    'NLD': 24,
    'POL': 25,
    'PRT': 26,
    'SVK': 27,
    'SVN': 28,
    'SWE': 29,
    'TUR': 30,
    'USA': 31,
    'ISR': 32,
    'ISL': 33,
    'ARG': 34,
    'AUT': 35,
    'IDN': 36,
    'NOR': 37,
    'BRA': 38,
    'CHL': 39,
    'CHN': 40,
    'G20': 41,
    'RUS': 42,
    'ZAF': 43,
    'IND': 44,
    'COL': 45,
    'NZL': 46,
    'SAU': 47
}
```

```python
# Custom Encoding on country column using countries dictionary
edu_attain["country_num"] = edu_attain["country"].apply(lambda x: countries[x])
edu_attain
```

```python
# Drop country column
edu_attain = edu_attain.drop(['country'], axis = 1)
edu_attain
```

```python
# Define features
x = edu_attain.copy()
x = x.drop("subject", axis = 1)
x.head()
```

```python
# Define the target set
y = edu_attain['subject'].ravel()
y[:12]
```

```python
# Split into Train & Test Sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 1, stratify = y)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Import Models
from sklearn import ensemble
from sklearn import naive_bayes
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
ml_list = {
    "LogisticRegression":sklearn.linear_model.LogisticRegression(solver='lbfgs', max_iter=200, random_state=1),
    "RandomForestClassifier": ensemble.RandomForestClassifier(),
    "GaussianNB": naive_bayes.GaussianNB(),
    "SVM": SVC(kernel = 'linear')
}
```

```python
results = []
for model in ml_list:
    ml = ml_list[model]
    ml.fit(x_train, y_train)
    y_pred = ml.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    results.append({
        "name": model,
        "accuracy":accuracy
    })
    print(f"{model} - Accuracy: {accuracy}")
```

```python
results_df = pd.DataFrame(results).sort_values("accuracy", ascending=False)
results_df
```

```python
# Making predictions using the testing data
rf_model = ensemble.RandomForestClassifier()
rf_model = rf_model.fit(x_train, y_train)
predictions = rf_model.predict(x_test)
```

```python
# Calculating the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1", "Actual 2"], 
    columns=["Predicted 0", "Predicted 1", "Predicted 2"]
)

cm_df
```

```python
# Accuracy Score
acc_score = accuracy_score(y_test, predictions)
acc_score
```

```python
# Display Classification Report
print("Classification Report")
print(classification_report(y_test, predictions))
```
